```python
import torch
from torch import nn
import torchvision
import torchvision.datasets as datasets
from torchvision import transforms as T
import matplotlib.pyplot as plt
import numpy as np
import os

data_dir = os.path.abspath(os.path.join("Project", "tiny-imagenet-200"))
transform = T.Compose([T.Resize(299), T.ToTensor(),])
dataset = datasets.ImageFolder(os.path.join(data_dir, "train"), transform)
dataloader = torch.utils.data.DataLoader(dataset, batch_size=32, num_workers=4, shuffle=True)

def matplotlib_imshow(img, one_channel=False):
    if one_channel:
        img = img.mean(dim=0)
    img = img / 2 + 0.5     # unnormalize
    npimg = img.numpy()
    if one_channel:
        plt.imshow(npimg, cmap="Greys")
        plt.imsave("tiny_collage.jpeg", npimg, cmap="Greys")
    else:
        plt.imshow(np.transpose(npimg, (1, 2, 0)))
        plt.imsave("tiny_collage.jpeg", np.transpose(npimg, (1, 2, 0)))
        
dataiter = iter(dataloader)
imgs, labels = next(dataiter)
img_grid= torchvision.utils.make_grid(imgs)
matplotlib_imshow(img_grid)
```
